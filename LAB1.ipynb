```python

def extract_text_from_file(file_path):

    if file_path.endswith((".doc", ".docx")):
        text = docx2txt.process(file_path)

    elif file_path.endswith(".pptx"):
        prs = Presentation(file_path)
        text = ""
        for slide in prs.slides:
            for shape in slide.shapes:
                if hasattr(shape, "text"):
                    text += shape.text + "\n"


    elif file_path.endswith(".pdf"):
        with pdfplumber.open(file_path) as pdf:
            text = ""
            for page in pdf.pages:
                text += page.extract_text() + "\n"

    elif file_path.endswith(".html"):
        with open(file_path, "r") as file:
            html_content = file.read()
            soup = BeautifulSoup(html_content, "html.parser")
            text = soup.get_text()


    else:
        text = ""

    return text


file_path = "FilGoal _ فى الجول _ الموقع الرياضى الرائد فى مصر و الوطن العربى.html"
extracted_text = extract_text_from_file(file_path)
print(extracted_text)
```

```python
def cvtojson(extracted_text):
  return {
      'data': extracted_text
  }
```

```python
extension = file_path.split(".")[-1]
print(extension)

data = cvtojson(extracted_text)
print(data)
```

```python
def normlization(path, extension, data):
  return {
      'path':path,
      'extension':extension,
      'data': extracted_text
  }

normlized_data = normlization(file_path, extension, data)
normlized_data
```

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")

def answer(context, question):
  result = qa_pipeline(question=question, context=context)
  return result['answer']

```

```python
question = "ما هي  نتيجه مبارة اليوم "
answer(normlized_data['data'], question)
```
